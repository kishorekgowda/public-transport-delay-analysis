```python
#  Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from datetime import datetime

#  Load the dataset
# Replace 'public_transport_delay.csv' with your actual file path
df = pd.read_csv('public_transport_delay.csv')

#  Data preprocessing
df['Date'] = pd.to_datetime(df['Date'])
df['Hour'] = df['Date'].dt.hour
df['Day'] = df['Date'].dt.date

# Set style for plots
sns.set(style="whitegrid")

#  Daily Average Delay Plot
daily_avg = df.groupby('Day')['Delay Duration'].mean()
plt.figure(figsize=(10, 5))
daily_avg.plot(marker='o')
plt.title('Average Daily Delay Duration')
plt.xlabel('Date')
plt.ylabel('Delay (minutes)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.savefig('daily_delay.png')
plt.show()

#  Hourly Delay Pattern
hourly_avg = df.groupby('Hour')['Delay Duration'].mean()
plt.figure(figsize=(10, 5))
sns.barplot(x=hourly_avg.index, y=hourly_avg.values, palette='Blues_d')
plt.title('Average Delay by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Delay (minutes)')
plt.tight_layout()
plt.savefig('hourly_delay.png')
plt.show()

#  Route-Based Delay Analysis
route_avg = df.groupby('Route')['Delay Duration'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=route_avg.index, y=route_avg.values, palette='Greens')
plt.title('Average Delay by Route')
plt.xlabel('Route')
plt.ylabel('Delay (minutes)')
plt.tight_layout()
plt.savefig('route_delay.png')
plt.show()

#  Pie Chart - Delay Contribution by Route
plt.figure(figsize=(8, 8))
route_delay_sum = df.groupby('Route')['Delay Duration'].sum()
plt.pie(route_delay_sum, labels=route_delay_sum.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Proportion of Total Delay by Route')
plt.tight_layout()
plt.savefig('route_delay_pie.png')
plt.show()


#  Geospatial Analysis (Optional - requires Jupyter or HTML output)
# Create a sample map with 100 random entries (or full if small)
sample_df = df.sample(min(len(df), 100))

map_center = [sample_df['Latitude'].mean(), sample_df['Longitude'].mean()]
delay_map = folium.Map(location=map_center, zoom_start=12)

for _, row in sample_df.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=3 + row['Delay Duration'] * 0.2,
        popup=f"Route: {row['Route']}, Delay: {row['Delay Duration']} min",
        color='red',
        fill=True,
        fill_opacity=0.6
    ).add_to(delay_map)

# Save the map
delay_map.save('delay_map.html')
```
